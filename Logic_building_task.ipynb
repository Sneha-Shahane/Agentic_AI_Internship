```python
# 2. Pass / Fail Analyzer
# Problem Statement:
# Analyze student results.
# Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Checking each student's marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Final result
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)

```

```python
# 3. Simple Data Cleaner
# Problem Statement:
# Clean and standardize user names.
# Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()   # remove spaces + lowercase
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

```python
#4. Message Length Analyzer
# Problem Statement:
# Analyze message sizes.


messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print(" Long message detected")
```

```python
# 5. Error Message Detector
# Problem Statement:
# Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
