```python
# Problem 1: Employee Performance Bonus Eligibility

# Employee performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find all employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Output
print("Top Performers Eligible for Bonus:", ", ".join(top_performers),
      f"(Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))

# Split into words
words = query.split()

# Count frequency
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

# Show words searched more than once
result = {word: count for word, count in freq.items() if count > 1}

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:   # even readings are valid
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_users = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average_sales = sum(sales) / len(sales)

# Threshold = 30% above average
threshold = average_sales * 1.3

# Detect spikes
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_ids = {}

# Count occurrences
for user in user_ids:
    count_ids[user] = count_ids.get(user, 0) + 1

# Display duplicates
for user, count in count_ids.items():
    if count > 1:
        print(f"{user} → {count} times")
```
