```python
# Problem 1: Smart Light Controller
# Problem Statement:
# Create a class to control a smart light that can be turned ON or OFF
# and display its status.
# Real-Time Use: Smart home automation systems

class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()

# Expected Output
light.display_status()
```

```python
# Problem 2: Employee ID Card System
# Problem Statement:
# Create a class that stores employee name, ID, and department
# and displays ID card details.
# Real-Time Use: Corporate HR systems

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Sample Input
emp = Employee("Rahul", "EMP102", "IT")

# Expected Output
emp.display_id_card()
```

```python
# Problem 3: Mobile Contact Record
# Problem Statement:
# Create a class to store contact name and phone number
# and display contact information.
# Real-Time Use: Mobile contact applications

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Sample Input
contact = Contact("Anita", "9876543210")

# Expected Output
contact.display_contact()
```

```python
# Problem 4: Product Price Tag Generator
# Problem Statement:
# Create a class that stores product name and price
# and prints a formatted price tag.
# Real-Time Use: Retail billing systems

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


# Sample Input
product = Product("Headphones", 2499)

# Expected Output
product.print_price_tag()
```

```python
# Problem 5: Movie Rating Display System
# Problem Statement:
# Create a class that stores movie name and rating
# and displays movie details.
# Real-Time Use: Streaming platforms

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


# Sample Input
movie = Movie("Inception", 4.8)

# Expected Output
movie.display_movie()
```

```python
# Problem 6: Delivery Address Manager
# Problem Statement:
# Create a class that stores customer name and delivery address
# and prints delivery details.
# Real-Time Use: Courier and logistics platforms

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Sample Input
delivery = Delivery("Suman", "Hyderabad")

# Expected Output
delivery.display_delivery_details()
```
